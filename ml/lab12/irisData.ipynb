```python
from __future__ import division

import csv

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = []
data_classes = []
with open('irisData.csv', 'rb') as csv_file:
    iris_data_reader = csv.reader(csv_file, delimiter=',', quotechar='|')
    for row in iris_data_reader:
        if row[4] == 'Iris-setosa':
            data.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])
            data_classes.append(1)
        elif row[4] == 'Iris-versicolor':
            data.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])
            data_classes.append(2)
        else:
            data.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])
            data_classes.append(3)
```

```python
data_tsne = TSNE(n_components=2, perplexity=50).fit_transform(data)
```

```python
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=data_classes, cmap=plt.cm.get_cmap("jet", 3))
plt.colorbar(ticks=range(4))
plt.show()

```

```python
pca = PCA(n_components=2, whiten=True)
pca.fit(data)
data_pca = pca.transform(data) 
```

```python
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data_classes, cmap=plt.cm.get_cmap("jet", 3))
plt.colorbar(ticks=range(4))
plt.show()
```

Больше нравится результат полученный алгоритмом t-sne, так как объекты из одного класса расположены кучней. Ну и вообще как я понял преимущество t-sne, что он сохраняет дистанции между объектами.
